#### listas con funciones arbitro

```python
class Arbitro:
    def __init__(self, nombre, puesto, salario):
        # 3 Atributos
        self.nombre = nombre
        self.puesto = puesto
        self.salario = salario

    # --- FUNCIONES ---
    def agregar(self):
        return f"✅ [AGREGAR]: Árbitro {self.nombre} registrado con éxito."

    def leer(self):
        return f"📖 [LEER]: Nombre: {self.nombre} | Puesto: {self.puesto} | Salario: ${self.salario}"

    def buscar(self, busqueda):
        if busqueda.lower() in self.nombre.lower():
            return f"🔍 [BUSCAR]: Se encontró coincidencia: {self.nombre} es {self.puesto}."
        return None

    def modificar(self, nuevo_puesto, nuevo_salario):
        self.puesto = nuevo_puesto
        self.salario = nuevo_salario
        return f"📝 [MODIFICAR]: Datos de {self.nombre} actualizados."

    def borrar(self):
        return f"❌ [BORRAR]: El objeto de {self.nombre} ha sido marcado para eliminación."

# --- BLOQUE DE ENTRADA (INPUTS) ---
def ejecutar_sistema():
    print("--- REGISTRO DE 3 ÁRBITROS ---")
    lista_arbitros = []

    # Bucle para capturar 3 árbitros
    for i in range(1, 4):
        print(f"\nDatos del Árbitro {i}:")
        nom = input(f"Ingrese nombre: ")
        pue = input(f"Ingrese puesto: ")
        sal = input(f"Ingrese salario: ")

        # Creamos el objeto con los datos del input
        nuevo_arbitro = Arbitro(nom, pue, sal)
        lista_arbitros.append(nuevo_arbitro)
        print(nuevo_arbitro.agregar())

    # --- DEMOSTRACIÓN DE FUNCIONES CON LOS OBJETOS CREADOS ---
    print("\n" + "="*30)
    print("SISTEMA DE GESTIÓN ACTIVO")
    print("="*30)

    # 1. Función Listar/Leer
    print("\n--- LISTADO (Función Leer) ---")
    for arb in lista_arbitros:
        print(arb.leer())

    # 2. Función Buscar
    print("\n--- BÚSQUEDA ---")
    buscado = input("¿A quién deseas buscar? (Escribe el nombre): ")
    encontrado = False
    for arb in lista_arbitros:
        resultado = arb.buscar(buscado)
        if resultado:
            print(resultado)
            encontrado = True
    if not encontrado: print("No se encontró nadie con ese nombre.")

    # 3. Función Modificar (Ejemplo con el primer árbitro)
    print(f"\n--- MODIFICAR (Cambiando datos de {lista_arbitros[0].nombre}) ---")
    n_puesto = input("Nuevo puesto: ")
    n_salario = input("Nuevo salario: ")
    print(lista_arbitros[0].modificar(n_puesto, n_salario))
    print(lista_arbitros[0].leer())

    # 4. Función Borrar
    print("\n--- BORRAR (Simulación) ---")
    print(lista_arbitros[-1].borrar()) # Borra el último de la lista

# Ejecutar el programa
if __name__ == "__main__":
    ejecutar_sistema()
```
