#### listas arbitro con funciones

```python
# Nuestra base de datos (Lista de listas)
lista_arbitros = []

# --- FUNCIONES ---

def agregararbitro(nombre, puesto, salario):
    # Guardamos los 3 atributos en la lista
    lista_arbitros.append([nombre, puesto, salario])
    print(f"\n>>> He agregado a '{nombre}' a la lista.")

def leerarbitro():
    # Solo muestra los nombres registrados
    nombres = [arb[0] for arb in lista_arbitros]
    print(f"\n>>> LEYENDO DATOS... La lista de nombres es: {nombres}")

def listararbitro():
    # Listado con todos los detalles
    print(f"\n>>> LISTADO DETALLADO DE ÁRBITROS:")
    if not lista_arbitros:
        print("La lista está vacía.")
    else:
        for i, arb in enumerate(lista_arbitros, 1):
            print(f"{i}. Nombre: {arb[0]} | Puesto: {arb[1]} | Salario: {arb[2]}")

def buscararbitro(nombre):
    encontrado = False
    for arb in lista_arbitros:
        if arb[0].lower() == nombre.lower():
            print(f"\n>>> ¡Encontrado! {arb[0]} trabaja como {arb[1]} con un salario de {arb[2]}.")
            encontrado = True
            break
    if not encontrado:
        print(f"\n>>> '{nombre}' no se encuentra en la base de datos.")

def modificararbitro(antiguo):
    encontrado = False
    for arb in lista_arbitros:
        if arb[0].lower() == antiguo.lower():
            print(f"\n>>> Modificando datos de {arb[0]}:")
            arb[0] = input("Nuevo nombre: ")
            arb[1] = input("Nuevo puesto: ")
            arb[2] = input("Nuevo salario: ")
            print(">>> Modificación exitosa.")
            encontrado = True
            break
    if not encontrado:
        print(f"\n>>> Error al modificar. '{antiguo}' no existe.")

def borrararbitro(nombre):
    for arb in lista_arbitros:
        if arb[0].lower() == nombre.lower():
            lista_arbitros.remove(arb)
            print(f"\n>>> Se ha eliminado a '{nombre}' de la lista.")
            return
    print(f"\n>>> Error al borrar. '{nombre}' no existe.")

# --- MENÚ INTERACTIVO ---

def menu_principal():
    print("--- SISTEMA DE GESTIÓN DE ÁRBITROS ---")

    while True:
        print("\n" + "—"*40)
        print(" MENU DE OPCIONES")
        print("—"*40)
        print("1. Agregar árbitro")
        print("2. Leer árbitros (Nombres)")
        print("3. Listar árbitros (Todo)")
        print("4. Buscar árbitro")
        print("5. Modificar árbitro")
        print("6. Borrar árbitro")
        print("7. Salir")

        opcion = input("\nSeleccione una acción (1-7): ")

        if opcion == "1":
            nom = input("Nombre del árbitro: ")
            pue = input("Puesto: ")
            sal = input("Salario: ")
            agregararbitro(nom, pue, sal)

        elif opcion == "2":
            leerarbitro()

        elif opcion == "3":
            listararbitro()

        elif opcion == "4":
            nombre = input("Nombre a buscar: ")
            buscararbitro(nombre)

        elif opcion == "5":
            antiguo = input("Nombre del árbitro a modificar: ")
            modificararbitro(antiguo)

        elif opcion == "6":
            nombre = input("Nombre a eliminar: ")
            borrararbitro(nombre)

        elif opcion == "7":
            print("Cerrando sistema. ¡Hasta pronto!")
            break

        else:
            print("Opción no válida, intente de nuevo.")

# Ejecución del programa
if __name__ == "__main__":
    menu_principal()
    print("Programa Terminado")
    print("Creadora de software Nicole Robles NL = 46")
```
